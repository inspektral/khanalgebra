## linear combinations

Just scaling and summing together vectors, no other operations happen

**Span** is what we can cover with a linear cominantions of two vectors, eg $span(\vec{a}, \vec{b}) = \mathbb{R^2}$, more formally:

$span(v_1, v_2, ... , v_n) = \{c_1v_1+c_2v_2 + ... + c_nv_n | c_i \in \mathbb{R}, 1 \leq i \leq n \}$

```python
a = np.array([[1], [2]])
b = np.array([[3], [0]])
c = np.array([[1], [-1]])

combination = 2*a - 0.5*b + 1.2*c

vu.plot_vector(a)
vu.plot_vector(b)
vu.plot_vector(c)
vu.plot_vector(combination, color='b')

vu.showPlot()
```

## linear independence

if we have $v_1 = cv_2$ they are linearly dependent as they lie on the same line, and $span(v_1, v_2) = span(v_1) = span(v_2)$, which is the line.

Formally a set is linearly dependent if an element is a linear combination of some other elements in the set, otherwise it is linearly independent.
