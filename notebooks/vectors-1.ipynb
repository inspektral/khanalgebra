```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vector(v, ref=(0,0), label="vector", color='r') :
    plt.quiver(ref[0], ref[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
```

```python
def showPlot(xlim=(-5, 5), ylim=(-5, 5), legend=True):
    
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)

    plt.xlim(*xlim)
    plt.ylim(*ylim)

    # Add grid, legend, and labels
    plt.grid()
    if legend:
        plt.legend()
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Vector Plot')
    plt.show()
```

```python
v = np.array([2,1])

plt.figure()
plot_vector(v)
showPlot()
```

```python
v_2 = v*2
v_3 = -1*v

plot_vector(v_2, color='b')
plot_vector(v_3, color='g')
plot_vector(v)
showPlot()
```

```python
a = np.array([-1,2])
b = np.array([3,1])
c = a+b

plot_vector(a, label="a")
plot_vector(b, ref=a, color='c', label="b")
plot_vector(b, color='b', label="b")
plot_vector(c, color='g', label="a+b")
showPlot()
```

```python
c = a-b
d = b-a

plot_vector(a, label="a")
plot_vector(b, color='b', label="b")
plot_vector(-b, color='c', label="-b")
plot_vector(a, ref=-b, label='a')
plot_vector(c,ref=b, color='g', label="a-b")
plot_vector(d,ref=c, color='g', label="b-a")
showPlot()
```

## Unit vectors

```python
i = np.array([1,0])
j = np.array([0,1])

plot_vector(i)
plot_vector(j, color='b')

showPlot(xlim=(-2,2), ylim=(-2,2))

def find_unit_vectors(vector: np.array):
    length = np.sqrt(np.sum(vector**2))
    return vector/length
```

```python
def find_components_length_angle(length: float, angle: float, degrees: bool = True):
    if degrees:
        angle = angle/180 * np.pi
    x = length*np.cos(angle)
    y = length*np.sin(angle)
    return np.array([x,y])
```

```python
v = {"length" : 3, "angle": 310}
w = {"length" : 8, "angle": 190}

v = find_components_length_angle(**v)
w = find_components_length_angle(**w)

sum = v+w

plot_vector(v)
plot_vector(w)
plot_vector(sum, label='sum', color='b')
showPlot(xlim=(-9,7), ylim=(-7,7))

print(sum)
```

```python
v = np.array([2,1])

for _ in range(30):
    w = v*(10*np.random.ranf()-5)
    plot_vector(w)

plot_vector(v,color='b')

showPlot(xlim=(-10,10), legend=False)
```

```python
v = np.array([[2], [1]])
plot_vector(v, color='b')
showPlot(xlim=(-10,10), ylim=(-10,10), legend=False)

def random_vectors(size, number=10, spread=5):
    rand = (spread*2)*np.random.ranf(number)-spread
    rand = np.ones(shape=(size, rand.size))*rand
    return rand

rand = random_vectors(v.size)

w = v*rand

for i in range(w.shape[1]):
    plot_vector(w[:, i])

plot_vector(v, color='b')
showPlot(xlim=(-10,10), ylim=(-10,10), legend=False) 


shift = np.array([[1], [3]])
w_shift = w+shift

for i in range(w.shape[1]):
    plot_vector(w[:, i])
plot_vector(v, color='b')
plot_vector(shift, color='m')
showPlot(xlim=(-10,10), ylim=(-10,10), legend=False) 

for i in range(w.shape[1]):
    plot_vector(w[:, i])
    plot_vector(shift, ref=w[:,i], color='c')
plot_vector(v, color='b')
plot_vector(shift, color='m')
showPlot(xlim=(-10,10), ylim=(-10,10), legend=False)

for i in range(w.shape[1]):
    plot_vector(w[:, i])
    plot_vector(shift, ref=w[:,i], color='c')
    plot_vector(w_shift[:, i], color='g')
plot_vector(v, color='b')
plot_vector(shift, color='m')
showPlot(xlim=(-10,10), ylim=(-10,10), legend=False)

for i in range(w.shape[1]):
    plot_vector(w[:, i])
    plot_vector(shift, ref=w[:,i], color='c')
    plot_vector(w_shift[:, i], color='g')
    plot_vector(w[:, i], ref=shift, color='y')
plot_vector(v, color='b')
plot_vector(shift, color='m')
showPlot(xlim=(-10,10), ylim=(-10,10), legend=False) 
```

## Find a line give two vectors

```python
a = np.array([[2],[1]])
b = np.array([[0],[3]])

diff = b-a

plot_vector(a)
plot_vector(b)
plot_vector(diff, ref=a, color='b', label='difference')
showPlot(legend=True)

rand = random_vectors(size=diff.size, spread=5)

line = b+(diff*rand)

for i in range(line.shape[1]):
    plot_vector(line[:,i])

plot_vector(a, color='b')
plot_vector(b, color='b')

showPlot(xlim=(-10,10), ylim=(-8,12), legend=False)
    
```
